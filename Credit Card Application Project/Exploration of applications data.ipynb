```python
def plotfield(column,kind='bar',title = '', ylabel = '',xlabel = '', figsize = (20,10), annotate = False, topn = 15,logscale = False,rotation = 0):
    sns.set(font_scale=2)
    sns.set_style("white")
    plt.figure(figsize = figsize)
    ax = data[column].value_counts().head(topn).plot(kind='bar')
    plt.xticks(rotation = rotation)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    if logscale:
        plt.yscale('log')
    if annotate:
        for bar in ax.patches:
            ax.annotate(format(bar.get_height(), '.0f'),
                           (bar.get_x() + bar.get_width() / 2,
                            bar.get_height()), ha='center', va='center',
                           size=20, xytext=(0, 8),
                           textcoords='offset points')
    
    
    return ax
```

```python
plotfield('lastname',annotate = False,title = 'Top 15 last names',ylabel = 'Count of last names',xlabel = 'Last names',rotation = 90)
```
